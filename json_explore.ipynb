```python
import json

# 1) Load the JSON dump
with open("all_labels.json", "r") as f:
    data = json.load(f)

# 2) Flatten into a single list of route records
routes = []
for source, lbls in data.items():
    for lbl in lbls:
        routes.append({
            "source": source,
            "cost": lbl["cost"],
            "dist_since_charge": lbl["dist_since_charge"],
            "time": lbl["time"],
            "current_time": lbl["current_time"],
            "path": lbl["path"]
        })

# 3) Sort by reduced cost ascending (i.e. most negative first)
routes_sorted = sorted(routes, key=lambda x: x["cost"])

# 4) Take top 10
top10 = routes_sorted[:10]

# 5) Print them
for i, r in enumerate(top10, start=1):
    print(f"{i}. Depot: {r['source']} → cost: {r['cost']:.2f}")
    print(f"   Path: {' → '.join(r['path'])}")
    print(f"   Distance since charge: {r['dist_since_charge']:.1f} km")
    print(f"   Cumulative time: {r['time']:.1f} min")
    print(f"   Arrival time: {r['current_time']}\n")
```
